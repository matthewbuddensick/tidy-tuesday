## Most Popular Seasons

```python
ordered_season_rank = summary[["season_name", "rank"]].sort_values("rank", ascending = True).dropna()

plt.figure(figsize = (12, 12))
sns.barplot(x = "rank", y = "season_name", data = ordered_season_rank, color = "lightblue", alpha = 0.8)
plt.xlabel("Rank")
plt.ylabel("Season")
plt.tick_params(axis = "y", left = False)
plt.title("Season Ranking by Viewers")
plt.show()
```

```python
fig, [ax1, ax2] = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 12))

sns.barplot(x = "viewers_premier", y = "season_name", data = summary, order = summary.sort_values("viewers_premier", ascending = False)["season_name"], color = "lightblue", alpha = 0.8, ax = ax1)
ax1.set_xlabel("Viewers (millions)")
ax1.set_ylabel("Season")
ax1.tick_params(axis = "y", left = False)
ax1.set_title("Premier Viewers of a Season")

sns.barplot(x = "viewers_finale", y = "season_name", data = summary, order = summary.sort_values("viewers_premier", ascending = False)["season_name"], color = "lightblue", alpha = 0.8, ax = ax2)
ax2.set_xlabel("Viewers (millions)")
ax2.set_ylabel("")
ax2.set_yticklabels("", minor =  False)
ax2.tick_params(axis = "y", left = False)
ax2.set_title("Finale Viewers of a Season")

plt.show()
```

```python
summary["difference_viewers"] = summary["viewers_finale"] - summary["viewers_premier"]

plt.figure(figsize = (8, 8))
sns.barplot(x = "difference_viewers", y = "season_name", data = summary, order = summary.sort_values("difference_viewers", ascending = False)["season_name"], color = "lightblue", alpha = 0.8)
plt.axvline(0, color = "black", alpha = 0.2)
plt.xlabel("Difference in Viewers (millions)")
plt.ylabel("Season")
plt.tick_params(axis = "y", left = False)
plt.title("Difference Between Finale and Premier Viewership")
plt.show()
```

Some seasons had drastically different premier and finale viewer numbers. Survivor: Borneo had relatively average premiere viewers but had the most viewers in the finale by far, it is also ranked as the second highest season among viewers. This drastic change from a decently watched season at the beginning to being the most watched final shows that viewers really enjoyed this season. The highest ranked season, Survivor: The Australian Outback, had the most viewers during the premier but viewership declined in the finale. Since Survivor: Borneo saw such a large increase in viewers from the premier to the finale, it is likely that the reunion episode also saw the highest or one of the highest viewer counts.

```python
# the seasons with the top 5 highest viewers for the reunion episode
summary.sort_values("viewers_reunion", ascending = False)[["season_name", "viewers_reunion"]][:5]
```

```python
# the first 5 seasons of the show
summary.sort_values("season", ascending = True)["season_name"][:5]
```

Survivor: Borneo had the highest viewer numbers out of all seasons during the reunion episode. It also happened to be the first season of the show, which means that the large increase in viewers from premier to finale is due to the new show gaining popularity very quickly. The second season of the show also happens to be Survivor: The Austalian Outback, the top ranked season, and it saw the highest viewer numbers for the premier. The show was able to continue its popularity after the season 1 success of Survivor: Borneo, but was not able to keep the same level of success as viewer numbers dropped as season 2 of the show went on.

```python
top_2_seasons = viewers[viewers["season_name"].isin(["Survivor: Borneo", "Survivor: The Australian Outback"])]

plt.figure(figsize = (8,8))
sns.lineplot(x = "episode", y = "viewers", hue = "season_name", data = top_2_seasons)
plt.xlabel("Episode Number")
plt.ylabel("Viewers (millions)")
plt.title("Viewer Counts Per Episode for Top Two Seasons")
plt.legend(title = "Season", loc = "lower right")
plt.xticks(list(range(1, 17)))
plt.show()
```

Survival: Borneo saw steady increase in viewers each episode and saw a huge spike from episode 12 to 13 and then a decrease again for the finale. Survival: The Australian Outback started out strong with similar viewership to the last episode of Survival: Borneo, but quickly saw viewership decrease until the last two episodes of the season where viewership spiked again.

Califronia has two of the top 3 cities that participants are chosen from, with over 80 of the 594 participants coming from Los Angeles and Santa Monica alone. New York City makes up over half of the participants chosen from New York as a whole with New York City and Brooklyn combining for 40 of the states 68 participants. 

#### Age

```python
age_bin_counts = pd.DataFrame(pd.cut(castaways["age"], bins = 8, precision = 0).value_counts()).reset_index().rename(columns = {"index": "age_group", "age": "count"})
age_bin_counts = age_bin_counts.astype({"age_group": "string"})
age_bin_counts
```

```python
plt.figure(figsize = (8, 8))
sns.barplot(x = "count", y = "age_group", data = age_bin_counts, order = ["(18.0, 25.0]", "(25.0, 32.0]", "(32.0, 39.0]", "(46.0, 54.0]", "(54.0, 61.0]", "(61.0, 68.0]", "(68.0, 75.0]"])
plt.xlabel("Frequency")
plt.ylabel("Age Group")
plt.title("Ages of Paricipants on Survivor")
plt.show()
```

A majority of the participants on the Survivor TV show are under the age of 32. However, there are a handful of participants that have been over the age of 50.

Viewership has steadily declined over the 40 seasons Survivor has been on TV.

## Conclusion

The TV show Survivor started off as a hit with season 1, Survivor: Borneo seeing an increase of over 30 million viewers from the premier episode to the finale, the largest positive change from premier to final viewership in the shows history. However, that success was short lived and with season 2 of the show, Survivor: The Australian Outback, viewership numbers decreased fast, and the season finished with the largest negative change in viewership from the premier to the final in the shows history. Viewership has been steadily decreasing over time. Up until about episode 100 of the show, there were about 20 million viewers per episode, now, almost at episode 600, the show is averaging less than 10 million per episode. Season 41 of Survivor is set to premier later this year after being delayed due to the COVID-19 pandemic.
